```python
# 8 - Multiple Line INSERT from a CSV file into a Relational Table - Fill the Table with Data

import mysql.connector

mydb_connection = mysql.connector.connect(
  host="localhost",
  user="root",
  password="password",
  database="EmployeeDB"
)

mycursor = mydb_connection.cursor()

for row in emp_df.itertuples():
            mycursor.execute('''INSERT INTO EmployeeDB.emp (SIN, Fname, Lname, Gender, Age)
                            VALUES (%s, %s, %s, %s, %s)
                            ''', 
                          (row.SIN,
                          row.Fname, 
                          row.Lname, 
                          row.Gender,
                          row.Age)
                          )
        
mydb_connection.commit()                       # Permanently save the new data rows into the table 

print(mycursor.rowcount, "record inserted.")
```
