```python
# 1. Program to print factorial of a number recursively. 
  
# Recursive function 
def recursive_factorial(n):   
    if n == 1:   
        return n   
    else:   
        return n * recursive_factorial(n-1)   

# user input 
num = int((input("Please input the number for the recursive factorial:")))

# check if the input is valid or not 
if num < 0:   
    print("Invalid input ! Please enter a positive number.")   
elif num == 0:   
    print("Factorial of number 0 is 1")   
else:   
    print("Factorial of number", num, "is:", recursive_factorial(num))
```

```python
# 2. Program to print the fibonacci series upto n_terms 
  
# Recursive function 
def recursive_fibonacci(n): 
    if n <= 1: 
        return n 
    else: 
        return(recursive_fibonacci(n-1) + recursive_fibonacci(n-2)) 

# user input
n_terms = int((input("Please input the number of terms for the recursive fibonacci:")))

# check if the number of terms is valid 
if n_terms <= 0: 
    print("Invalid input ! Please input a positive value") 
else: 
    print("Fibonacci series:") 
    for i in range(n_terms): 
        print(recursive_fibonacci(i)) 
```

```python
# 4. Search Binary Tree Nodes  

class TreeNode:
    def __init__(self, dataval, left=None, right=None, parent=None):

        self.left = None
        self.right = None
        self.data = dataval
        self.parent = parent


    # findval method to compare the value with nodes
    def searchTree(self, searchVal):

        if self is not None:
            if searchVal < self.data:
                if self.left is None:
                    return str(searchVal)+" Not Found"
                return self.left.searchTree(searchVal)

            elif searchVal > self.data:
                if self.right is None:
                    return str(searchVal)+" Not Found"
                return self.right.searchTree(searchVal)

            else:
                print(str(self.data) + ' is found')
        else:
            print("Tree is empty")


# a. Driver code to build tree of root and other nodes
if __name__ == '__main__':
    root = TreeNode(20)
    root.left = TreeNode(10)
    root.left.left = TreeNode (5)
    root.left.left.left = TreeNode(12)
    root.left.left.right = TreeNode(7)
    root.right = TreeNode(30)
    root.right.left = TreeNode(25)
    root.right.right = TreeNode(35)
    root.right.right.right = TreeNode(100)

    # b. Driver code to search for a node in the tree
    print(root.searchTree(5))
    print(root.searchTree(48))
    print(root.searchTree(20))
```

```python
# 5. Check for Valid BST

# Data structure to store a Binary Search Tree node
class TreeNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right
    
# Function to print binary tree. It does reverse inorder traversal  
def printTreeNodes(root, level=0): 

        # Base case  
    if (root == None): 
        return

        # Process RIGHT child first  
    #if root.right != None:
    printTreeNodes(root.right, level+4)  

            # Print current node after level count  
    print()  
    for i in range(level): 
        print(end = " ")
    print(root.data)

        # Process LEFT child
    #if root.left != None:
    printTreeNodes(root.left, level+4)  

    # Print wrapper over the nodes 
def printTreeRoot(root): 
        
        # Pass initial level as 0  
    printTreeNodes(root, 0)  

    
# Function to determine if given Binary Tree is a BST or not by keeping a
# valid range (starting from [-INFINITY, INFINITY]) and keep shrinking
# it down for each node as we go down recursively
def isBST(node, minKey, maxKey):

    # base case
    if node is None:
        return True

    # if node's value fall outside valid range
    if node.data < minKey or node.data > maxKey:
        return False

    # recursively check left and right subtrees with updated range
    return isBST(node.left, minKey, node.data) and \
            isBST(node.right, node.data, maxKey)


# Function to determine if given Binary Tree is a BST or not
def checkForBST(root):

    if isBST(root, float('-inf'), float('inf')):
        print("This is a BST.")
    else:
        print("This is NOT a BST")


# a. Driver code for building tree of root and other nodes
if __name__ == '__main__':
    root = TreeNode(20)

    root.left = TreeNode(10)
    root.left.left = TreeNode(5)
    root.left.right = TreeNode(12) 
    root.left.left.left = TreeNode(7)
    root.left.left.right = TreeNode(7)

    root.right = TreeNode(30)
    root.right.left = TreeNode(28)
    root.right.right = TreeNode(45)
    #root.right.left.right = TreeNode(38)
    #root.right.left.right = TreeNode(100)

    # b Driver code for printing out the BST
    printTreeRoot(root)

    # c. Driver code for checking for valid BST
    checkForBST(root)
```
