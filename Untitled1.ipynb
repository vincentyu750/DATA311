```python
#Vincent Yu T03 October 15 Assignment 02 v1.0
#Question 2 sortCSV

import csv

def mergeSort(list):
    if len(list) > 1:
        mid = len(list) // 2 
        left = list[:mid]
        right = list[mid:]
        
        print("left:\n"+str(left).replace("],","]\n")) 
        print("right:\n"+str(right).replace("],","]\n")) 
        print("")
           
        #recursion
        mergeSort(left)
        mergeSort(right)
        
        i = j = k = 0
        
        while i < len(left) and j < len(right):
            if left[i][0] < right[j][0]:
                list[k][0] = left[i][0]
                i += 1
            else:
                list[k][0] = right[j][0]
                j += 1
            k += 1

        while i < len(left):
            list[k][0] = left[i][0]
            i += 1
            k += 1

        while j < len(right):
            list[k][0] = right[j][0]
            j += 1
            k += 1


def sortCSV(f1,f2):
    csvList = []
    with open(f1) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter = ',')
        next(csv_reader, None)
        for row in csv_reader:
            csvList.append(row)

    mergeSort(csvList)
    print(str(csvList).replace("],","]\n"))        
    

sortCSV("test1.csv","output1.csv")
```
