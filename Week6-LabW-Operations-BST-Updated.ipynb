```python
# Insert new data, Search for data, Get minimum data value, Get height of BST, Print a BST

# Class for BST
class BST:
    
    # Data structure to store a Binary Search Tree node
    class TreeNode:
        def __init__(self, data=None, left=None, right=None):
            self.data = data
            self.left = left
            self.right = right

    
    # Self constructor for initialization
    def __init__(self, root=None):
        self.root = root
        
        
    # 1. Insert new data in BST
    def insert(self, newNodeData):    
        curNode = self.root
        
        if curNode is None:
            self.root = self.TreeNode(newNodeData)
            return
        
        while curNode is not None:
            
            if newNodeData < curNode.data:
                if curNode.left is None:
                    curNode.left = self.TreeNode(newNodeData)
                    return
                else:
                    curNode = curNode.left
                    
            elif newNodeData > curNode.data:
                if curNode.right is None:
                    curNode.right = self.TreeNode(newNodeData)
                    return
                else:
                    curNode = curNode.right
    
    
    # 2. Search for data in BST
    def search(self, searchData):
        
        curNode = self.root
        
        while curNode is not None:
            
            if searchData < curNode.data:
                curNode = curNode.left
                
            elif searchData > curNode.data:
                curNode = curNode.right
                
            else:
                return "Node data " + str(curNode.data) + ' is found'
        
        if curNode is None:
            return "Node data " + str(searchData) + " is NOT Found"
        
    # 3. Get minimum data value in BST
    def get_min_value(self):
    
        curNode = self.root
  
        # loop down to find the lefmost leaf 
        while(curNode.left is not None): 
            curNode = curNode.left 

        return "The mininum data value is: " + str(curNode.data) 

    # 4. Get height of BST
    def get_tree_height(self):
        
        h = 0
        if self.root is not None:
            return "The height of the tree is: " + str(self.get_max_height(self.root, h))
        else: 
            return h

    def get_max_height(self, curNode, h):

        if curNode is None: return h
        
        left_height = self.get_max_height(curNode.left, h + 1)
        right_height = self.get_max_height(curNode.right, h + 1)
        
        return max(left_height, right_height)
    

    # 5. Print a BST
    def print(self):
        def printTree(n, indent = 0):
            
            if n.right is not None: 
                printTree( n.right, indent + 4)
                
            print( " " * indent, end = "")    
            print(n.data)
            
            if n.left is not None: 
                printTree( n.left, indent + 4)
                
        printTree(self.root)

    
# Driver code
if __name__ == '__main__':
    tree = BST()
    
    # a. Insert in the BST
    tree.insert(10)
    tree.insert(2)
    tree.insert(8)
    tree.insert(4)
    tree.insert(9)
    tree.insert(15)
    tree.insert(30)
    tree.insert(25)
    tree.insert(35)
    tree.insert(45)
    tree.insert(16)
    
    # b. Search for a node in BST
    print()
    print(tree.search(16))
    
    # c. Get minimum value of BST
    print()
    print(tree.get_min_value())

    # d. Get height of BST
    print()
    print(tree.get_tree_height())
    
    # e. Driver code to print out the tree
    print()
    tree.print()

    
```
