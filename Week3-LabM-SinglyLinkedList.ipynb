```python
### - 6. Remove a Tail Node

class Node:
    def __init__(self, data=None, nextval=None):
        self.dataval = data
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

    # Function to remove node
    def removeTailNode(self):
        if self.headval is not None:
                     
            printval = self.headval
            while (printval):
                tailval = printval
                printval = printval.nextval
                                
            #tailval = None
            tailval.dataval = None
            
        else:
            print("The Linked List is Null")
            
    def listPrint(self):
        printval = self.headval
        while (printval):
            print(printval.dataval)
            printval = printval.nextval
            
            
list = SLinkedList()
list.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

list.headval.nextval = e2
e2.nextval = e3

print("Original Linked List")
list.listPrint()
print("\n")

list.removeTailNode()

print("New Linked List without Previous Tail")
list.listPrint()
```
