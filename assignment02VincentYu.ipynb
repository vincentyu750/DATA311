```python
#Vincent Yu T03 October 13 Assignment 02 v1.0
#Question 2 singly linked list

#singly linked list class from lectures / formatted for my inability to read
        
class Node:
     def __init__(self, val=None, next=None):
        self.val = val
        self.next = next    
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def countNode(self):
        currentNode=self.head
        #first calculates how long the list is iterativly 
        listCount = 0
        currentNode = self.head
        while currentNode is not None:
            listCount += 1
            currentNode = currentNode.next
        print(str(listCount)+" Elements")
        
    def printList(self):
        node=self.head
        nextNode = None
        #check if list is empty (when there is no head element)
        if self.head is None:
            print("Empty List")
            return
        
        #prints list if not empty
        print("[",end="")
        while node is not None:
            #if the current node is the tail, if the next node is none it won't print with a comma
            if node.next is None:
                print(node.val, end="")
            #otherwise prints as usual
            else:
                print(node.val, end=",")
            node = node.next
        print("]")
            
    def insertNode(self, data, pos):
        '''
        0 for in front of all the elements (head)
        if the position is the same as the length of the list adds it to the end (tail)
        if anywhere else inserts it behind the node in that position
        '''
        #first calculates how long the list is iterativly 
        listCount = 0
        posCount = 0
        currentNode = self.head
        while currentNode is not None:
            listCount += 1
            currentNode = currentNode.next
            
        if data is None:
            print("You cannot add nothing to the list, you have to add something")
            return
            
        #if they enter a negative number
        if pos < 0:
            print("The current position ["+str(pos)+"] is out of boundaries, the current boundaries are 0-"+str(listCount-1)+" therefore ["+str(data)+"] cannot be added to the list")
            return
        
        #if the position asked for is out of the list boundaries 
        if pos > listCount:
            print("The current position ["+str(pos)+"] is out of boundaries, the current boundaries are 0-"+str(listCount-1)+" therefore ["+str(data)+"] cannot be added to the list")
            return
        
        #if the position asked for is in the front
        if pos == 0:
            insertionNode = Node(data)
            insertionNode.next = self.head
            self.head = insertionNode 
            print("The Operation was sucessful")
            return
        
        #if the position asked for is at the end
        if pos == listCount:
            insertionNode = Node(data)
        
            #if head is empty
            if self.head is None:
                self.head = insertionNode
                print("The Operation was sucessful")
                return 
        
            currentNode = self.head
            while currentNode.next is not None:
                currentNode = currentNode.next
            currentNode.next = insertionNode
        
        #if the position asked for is in the middle
        else:
            pos = pos - 1
            positionNode = self.head
            #finds where the position node is according to which position it asked for
            while posCount < pos:
                posCount += 1
                positionNode = positionNode.next
            
            insertionNode = Node(data)
            insertionNode.next = positionNode.next
            positionNode.next = insertionNode
            print("The Operation was sucessful")

    def deleteNode(self, pos):
        #checks if the list is null / empty nothing to delete
        if self.head == None:
            print("This list is not initialized, there is currently nothing in the list")
            return

        #first calculates how long the list is iterativly 
        listCount = 0
        posCount = 0
        currentNode = self.head
        selectedNode = self.head
        while currentNode is not None:
            listCount += 1
            currentNode = currentNode.next
            
        #if they enter a negative number
        if pos < 0:
            print("The current position ["+str(pos)+"] is out of boundaries, the current boundaries are 0-"+str(listCount-1))
            return
        
        #if the position asked for is out of the list boundaries 
        if pos > listCount:
            print("The current position ["+str(pos)+"] is out of boundaries, the current boundaries are 0-"+str(listCount-1))
            return
    
        #if the node you want to delete is the head node
        if pos == 0:
            self.head = selectedNode.next
            print("The Operation was sucessful")
            return
        
    
        #if the node you want to delete is in the middle or the tail end
        else:
            pos = pos - 1
            selectedNode = self.head
            #finds where the position node is according to which position it asked for
            while posCount <= pos:
                previousNode = selectedNode
                selectedNode = selectedNode.next
                posCount += 1
                
            previousNode.next = selectedNode.next
            selectedNode = None
            print("The Operation was sucessful")

```

```python
#Vincent Yu T03 October 15 Assignment 02 v1.0
#Question 2 sortCSV

import csv

#using merge sort to sort the csv 
def mergeSort(mylist):
    if len(mylist) > 1:
        mid = len(mylist) // 2 
        left = mylist[:mid]
        right = mylist[mid:]
        
        #recursion
        mergeSort(left)
        mergeSort(right)
        
        i = j = k = 0
        
        while i < len(left) and j < len(right):
            if int(left[i][0]) < int(right[j][0]):
                mylist[k] = left[i]
                i += 1
            else:
                mylist[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            mylist[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            mylist[k] = right[j]
            j += 1
            k += 1


def sortCSV(f1,f2):
    csvList = []
    with open(f1) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter = ',')
        
        #keeping the header seperate from the data
        head = next(csv_reader)
        
        next(csv_reader, None)
        for row in csv_reader:
            csvList.append(row)
    
    #merge sorts the list based off first column
    mergeSort(csvList)

    with open(f2, 'w') as outputFile:
        outputFile.write(str(head).replace(" ","").replace("[","").replace("]","").replace("'","")+'\n')
        for row in csvList:
            outputFile.write(str(row).replace(" ","").replace("[","").replace("]","").replace("'","")+ '\n')

sortCSV("test1.csv","output1.csv")
```
