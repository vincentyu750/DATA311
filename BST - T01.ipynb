```python
"""
getHeight(node20)
getHeight(node10)
getHeight(node5)
0
getHeight(node7)
1
2
getHeight(node16)
1
3
getHeight(node20) (leftsubtree = 3)
getHeight(node30)
getHeight(node25)
1
getHeight(node35)
1
2
getHeight(node20) (rightsubtree = 2)

4
""" 


class TreeNode:
    def __init__(self, value, left = None, right= None, parent=None):
        self.value = value
        self.left = left
        self.right = right
        self.parent = parent  

        
    """
    This method prints out the values in the tree node using dfs
    Parameter: Treenode
    """
    def printTree(self, level = 0):
        if self.right != None:
            self.right.printTree(level+4)
        print(" "*level, end="")
        print(self.value)
        if self.left!=None:
            self.left.printTree(level+4)
        
    """
    This method inserts a node in a BST
    Parameter: Root: TreeNode, Node: New TreeNode
    """
    def insert(self, node):
        if self is None: 
            self = node
        else:
            #Check is root is greater than node
            if self.value > node.value:
            #If node is less, we move left
                if self.left is None:
                #If left of root is empty, add new node here
                    self.left = node
                else:
                    #If left of root is not empty, recursive call with new root node
                    self.left.insert(node)
            else:
            #if root is less than node value, we move right
            #check is right of root is empty, if empty, add neew node here
                if self.right is None:
                    self.right = node
                else:
                    
                    self.right.insert(node)
                    
                    
        
                 
    def findMinNode(self):
        #n = self
        while self.left is not None:
            self = self.left
        return self
    
    
    def getHeight(self):
        if self is None:
            return 0
        else:
            leftsubtree,rightsubtree = 0,0 
            if self.left is not None:
                leftsubtree = self.left.getHeight()
            if self.right is not None:
                rightsubtree = self.right.getHeight()
            return 1 + max(leftsubtree, rightsubtree)
"""
getHeight(30)
getHeight(25)
getHeight(25.left) = 0 (leftsubtree)
getHeight(26)
getHeight(26.left) = 0 (leftsubtree)
getHeight(26.right) = 0 (rightsubtree)
getHeight(26) = 1
getHeight(25.right) = 1 (rightsubtree)
getHeight(25) = 2 
getHeight(30.left) = 2 (leftsubtree)

getHeight(30.right)
getHeight(35)
getHeight(35.left) = 0 (left)
getHeight(35.right) = 0 (right)
getHeight(35) = 1 
getHeight(30.right) = 1 (rightsubtree)

3

"""    
      

        
def main():
    #Create tree
    node100 = TreeNode(100)
    node7 = TreeNode(7)
    node5 = TreeNode(5, None, node7)
    node35 = TreeNode(35, None)
    node25 = TreeNode(25)
    node30 = TreeNode(30, node25, node35)
    node10 = TreeNode(10, node5)
    node20 = TreeNode(20, node10, node30)
    
    #Parent Declarations
    node30.parent = node20
    node10.parent = node20
    node25.parent = node30
    node35.parent = node30 
    node100.parent = node35
    node5.parent = node10
    node7.parent = node5
    
    
    #node20.printTree()
    node30.insert(TreeNode(26))
    #insert(node20, TreeNode(34))
    node30.printTree()
    
    #node10.printTree()
    #print(node20.findMinNode().value)
    
    print(node30.getHeight())
    
    
    
main()
        
```
