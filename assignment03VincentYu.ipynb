```python
#Vincent Yu T03 October 13 Assignment 03 v1.0
#Question 3 BST

class BST:
    
    #code from tutorial Week6-LabW-Operations-BST-Updated by Mireku Kwakye, Michael
    class TreeNode:
        def __init__(self, data, left=None, right=None, parent=None):
            self.data = data
            self.left = left
            self.right = right
            self.parent = parent
            
    # Self constructor for initialization
    def __init__(self, root = None):
        self.root = root

        
    # 1. Insert new data in BST
    def insert(self, newNodeData):    
        curNode = self.root
        
        if curNode is None:
            self.root = self.TreeNode(newNodeData)
            return
        
        while curNode is not None:
            
            if newNodeData < curNode.data:
                if curNode.left is None:
                    curNode.left = self.TreeNode(newNodeData)
                    #setting the parent node for deletion
                    curNode.left.parent = curNode
                    return
                else:
                    curNode = curNode.left
                    
            elif newNodeData > curNode.data:
                if curNode.right is None:
                    curNode.right = self.TreeNode(newNodeData)
                    curNode.right.parent = curNode
                    return
                else:
                    curNode = curNode.right
    
    # 5. Print a BST
    def print(self):
        def printTree(n, indent = 0):
            
            if n.right is not None: 
                printTree( n.right, indent + 4)
                
            print( " " * indent, end = "")    
            print(n.data)
            
            if n.left is not None: 
                printTree( n.left, indent + 4)
                
        printTree(self.root)
        
    #finding max value within tree
    def findMax(self):
        curNode = self.root
        # loop down to find the right most leaf
        while(curNode.right is not None): 
            curNode = curNode.right 
        return "The maximum data value is: " + str(curNode.data)
    
    def find(self,dataKey):
        curNode = self.root
        if curNode is not None:
            return self.recursiveFind(dataKey,self.root)
        else:
            print("Unable to find")
            return None
    
    def recursiveFind(self,dataKey,curNode):
        if dataKey == curNode.data:
            return curNode
        elif dataKey < curNode.data and curNode.left is not None:
            return self.recursiveFind(dataKey,curNode.left)
        elif dataKey > curNode.data and curNode.right is not None:
            return self.recursiveFind(dataKey,curNode.right)
        
    
    #bst node deletion
    def delete(self,dataKey):
        print("Currently removing: Node value: "+str(dataKey))
        return self.deleteNode(self.find(dataKey))
    
    def deleteNode(self, curNode):
        #find min node helps with deletion
        def findMin(node):
            currentNode = node
            # loop down to find the lefmost node
            while(currentNode.left is not None): 
                currentNode = currentNode.left 
            return currentNode
        
        parentNode = curNode.parent
        #case 1 if there are no children
        if curNode.left is None and curNode.right is None:
            #checks the parent of the node selected to remove the refrence
            if curNode.parent.left == curNode:
                curNode.parent.left = None
            if curNode.parent.right == curNode:
                curNode.parent.right = None
                

        #case 2 if there are one children
        if curNode.left is not None and curNode.right is None or curNode.right is not None and curNode.left is None:
            #checks which one has the child
            #if its left
            if curNode.left is not None:
                child = curNode.left
            #if its right
            else:
                child = curNode.right
                
            #replacing the parent node and changing the child node of the parent node accordingly
            if curNode.parent.left == curNode:
                curNode.parent.left = child
            if curNode.parent.right == curNode:
                curNode.parent.right = child
                
            #filling in the gap created
            child.parent = curNode.parent
                
        #case 3 if there are 2 or more children
        if curNode.left is not None and curNode.right is not None:
            #find sucessor using the min value function created and use that value for what we need to delete
            sucessorNode = findMin(curNode.right)
            curNode.data = sucessorNode.data
            #start the recursion in order to complete the deletion
            self.deleteNode(sucessorNode)
            
    def sumOfValues(self,condition):
        tempTree = BST()
        #printing all the values under condition value
        print("The Values That are under "+str(condition)+" are: ",end='')
        node = self.root
        def printInorder(n):
            if n is None:
                return
            printInorder(n.left)
            if n.data < condition:
                print(n.data, end=' ')
                tempTree.insert(n.data)
            printInorder(n.right)
        printInorder(node)
        #counting the sum of those values
        def countSum(n):
            if n is None:
                return 0
            else:
                return (n.data + countSum(n.left) + countSum(n.right))
        sum = countSum(tempTree.root)
        print("\nTherefore the sum is: "+str(sum))
        
    #finding the number of nodes in a tree (This function has only the self parameter and return the number of node)
    def sizeBST(self):
        node=self.root
        def countNodes(n):
            if n is None:
                return 0
            else:
                return (1 + countNodes(n.left) + countNodes(n.right))
        sum = int(countNodes(node))
        return sum
        
    # Function to determine if given Binary Tree is a BST or not
    def checkForBST(self):
        nnode = self.root
        # Function to determine if given Binary Tree is a BST or not by keeping a
        # valid range (starting from [-INFINITY, INFINITY]) and keep shrinking
        # it down for each node as we go down recursively
        def checkingBST(node, minKey, maxKey):
            # base case
            if node is None:
                return True

            # if node's value fall outside valid range
            if node.data < minKey or node.data > maxKey:
                return False
            # recursively check left and right subtrees with updated range
            return checkingBST(node.left, minKey, node.data) and checkingBST(node.right, node.data, maxKey)

        if checkingBST(nnode,float('-inf'), float('inf')):
            print("This is a BST.")
        else:
            print("This is NOT a BST")
```
